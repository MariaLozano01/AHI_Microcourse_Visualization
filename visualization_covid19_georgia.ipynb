##Transforming Columns 

```python
data['DATESTAMP']
```

```python
# Creating a copy of the original column so we keep the original version 
# we could also override the column if we wanted to, but because we're unsure about where we're gonna keep 
#the analysis, let's just keep it

data['DATESTAMP_MOD'] = data['DATESTAMP'] #creates a duplicate of the column 'DATESTAMP'
print(data['DATESTAMP_MOD'].head(10)) #create a print of the first 10 columns 
print(data['DATESTAMP_MOD'].dtypes) #print the type of variables present within this column 
```

```python
data['DATESTAMP_MOD'] = pd.to_datetime(data['DATESTAMP_MOD']) ##transforms 'DATESTAMP_MOD' into a timestamp column 
#instead of an object dtype
data['DATESTAMP_MOD'].dtypes
```

```python
data[['DATESTAMP','DATESTAMP_MOD']] ##DATESTAMP_MOD is no longer a string value but a datetime64
```

```python
data['DATESTAMP_MOD_DAY'] = data['DATESTAMP_MOD'].dt.date # takes information from the 'DATESTAMP_MOD column 
#and extracts the date section
data['DATESTAMP_MOD_DAY'] #outputted only the day field
```

```python
data['DATESTAMP_MOD_YEAR'] = data['DATESTAMP_MOD'].dt.year
data['DATESTAMP_MOD_MONTH'] = data['DATESTAMP_MOD'].dt.month
```

```python
data['DATESTAMP_MOD_YEAR']
```

```python
data['DATESTAMP_MOD_MONTH']
```

```python
data
```

```python
data['DATESTAMP_MOD_MONTH_YEAR'] = data['DATESTAMP_MOD'].dt.to_period('M')
data['DATESTAMP_MOD_MONTH_YEAR'].sort_values() #we get the first 5 rows and the bottom 5 rows
```

```python
data['DATESTAMP_MOD_WEEK'] = data['DATESTAMP_MOD'].dt.week 
data['DATESTAMP_MOD_WEEK'] #we get information on the week that the data was captured 
```

```python
data['DATESTAMP_MOD_QUARTER'] = data['DATESTAMP_MOD'].dt.to_period('Q')
data['DATESTAMP_MOD_QUARTER'].sort_values()
```

```python
data['DATESTAMP_MOD_DAY_STRING'] = data['DATESTAMP_MOD_DAY'].astype(str) #creates a new column but
#it is now converted into a string
data['DATESTAMP_MOD_WEEK_STRING'] = data['DATESTAMP_MOD_WEEK'].astype(str)#creates a new column but
#it is now converted into a string
data['DATETIME_STRING'] = data['DATESTAMP_MOD_MONTH_YEAR'].astype(str)
data['DATESTAMP_MOD_MONTH_YEAR'] = data['DATESTAMP_MOD_MONTH_YEAR'].astype(str)
```

##Creating the final dataframe/ specific columns/features/attributes that we want

```python
finalDF = selectCountyTime_aprilmay2020[['COUNTY',
                                         'DATESTAMP_MOD',
                                         'DATESTAMP_MOD_DAY', 
                                         'DATESTAMP_MOD_DAY_STRING', 
                                         'DATETIME_STRING', 
                                         'DATESTAMP_MOD_MONTH_YEAR',
                                         'C_New', #cases - new
                                         'C_Cum', #cases - total
                                         'H_New', #Hospitalizations - new
                                         'H_Cum', #hospitalizations - total
                                         'D_New',
                                         'D_Cum'                        ]]
```

```python
dailySpecific['newHospandDeathCovid'] = dailySpecific['D_New'].astype(int) + dailySpecific['H_New'].astype(int) + dailySpecific['C_New'].astype(int)
dailySpecific['newHospandDeathCovid']
```
